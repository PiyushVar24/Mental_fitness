#Encoding the Categorical values

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'object':
        data[i]=l.fit_transform(data[i])


data.head()
```

```python
data.shape
```

```python
data['Country'].unique()
```

#Spliting data


```python
x = data.drop('mental_fitness',axis=1)
y = data['mental_fitness']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
x.shape
```

```python
y.shape
```

#Linear Regression


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import r2_score,mean_squared_error
print("Mean square Error={}\nR-square Score={}".format(mean_squared_error(y_test,y_pred), r2_score(y_test,y_pred)))
```

#SVM Regression

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)

y_pred=svr.predict(x_test)

from sklearn.metrics import r2_score,mean_squared_error
print("Mean Square Error={}\nR-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))
```

#Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=15,random_state=12)
rf.fit(x_train,y_train)

y_pred = rf.predict(x_test)

from sklearn.metrics import mean_squared_error,r2_score
print("Mean Square Error={}\nR-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))
```

#Decision Tree Regression


```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=12)
dtr.fit(x_train,y_train)

y_pred = dtr.predict(x_test)

from sklearn.metrics import mean_squared_error,r2_score
print("Mean Square Error={}\nR-Square Score={}".format(mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)))
```

#Conclusion:

Random Forest Regression works well on both train and test sets with r2 score of 0.992.

As well as Decision Tree Regression also works well on both train and test set with r2 score of 0.984.
